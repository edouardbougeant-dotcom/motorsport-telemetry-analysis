## Comparaison lap time between 1st and 2nd during Monaco Grand Prix Qualification 2023

The goal of this analysis is to compare the fastest lap of Max Verstappen (1st) 
and Alonso (2nd) during the 2023 Monaco Grand Prix Qualification.
We will compare speed, throttle and brake to understand the differences 
in driving style between the two drivers.

```python
import fastf1

session = fastf1.get_session(2023, 'Monaco', 'Q')
session.load()
```

```python
for driver in session.drivers:
    fastest = session.laps.pick_driver(driver).pick_fastest()
    driver_info = session.get_driver(driver)
    print(driver_info['Abbreviation'], fastest['LapTime'])
```

```python
fastest_lap_VER = session.laps.pick_driver('VER').pick_fastest()
fastest_lap_ALO = session.laps.pick_driver('ALO').pick_fastest()

print(f"Fastest lap of VER : {fastest_lap_VER['LapTime']}")
print(f"Fastest lap of ALO : {fastest_lap_ALO['LapTime']}")
```
