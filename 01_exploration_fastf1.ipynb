```python
import fastf1

# Charger la session Qualifications Monaco 2023
session = fastf1.get_session(2023, 'Monaco', 'Q')
session.load()

print(session)
```

```python
session.laps
```

```python
Fast_lap_VER = session.laps.pick_driver('VER').pick_fastest()
print(Fast_lap_VER)
```

```python
Fast_lap_VER = session.laps.pick_driver('VER').pick_fastest().get_telemetry()
print(Fast_lap_VER)
```

```python
import matplotlib.pyplot as plt

plt.plot(Fast_lap_VER['Distance'], Fast_lap_VER['Speed'])
plt.title('Evolution of speed over distance for Max Verstappen fastest lap in Monaco 2023 Qualificatio')
plt.xlabel('Distance (m)')
plt.ylabel('Speed (km/h)')
plt.show()
```

```python
print(Fast_lap_VER.columns)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 1, figsize=(20, 8), sharex=True)

axs[0].plot(Fast_lap_VER['Distance'], Fast_lap_VER['Speed'])
axs[0].set_ylabel('Speed (km/h)')

axs[1].plot(Fast_lap_VER['Distance'], Fast_lap_VER['Brake'], color='red')
axs[1].plot(Fast_lap_VER['Distance'], Fast_lap_VER['Throttle']/100, color='green')
axs[1].set_ylabel('Throttle in green (%) and Brake in red (%)')

axs[1].set_xlabel('Distance (m)')

plt.suptitle('Verstappen - Monaco 2023 Qualifying')
plt.show()
```
