## First hand on data analysis python tools on the Monaco Grand Prix. 

```python
import fastf1

# Charger la session Qualifications Monaco 2023
session = fastf1.get_session(2023, 'Monaco', 'Q')
session.load()

print(session)
```

```python
session.laps
```

```python
Fast_lap_VER = session.laps.pick_driver('VER').pick_fastest()
print(Fast_lap_VER)
```

## Getting data of Max Verstappen's pole position lap to set a goal of performance.

```python
Fast_lap_VER = session.laps.pick_driver('VER').pick_fastest().get_telemetry()
print(Fast_lap_VER)
```

## Visualisation of the monaco track through speed evolution over distance graphic.  

```python
import matplotlib.pyplot as plt

plt.plot(Fast_lap_VER['Distance'], Fast_lap_VER['Speed'])
plt.title('Evolution of speed over distance for Max Verstappen fastest lap in Monaco 2023 Qualificatio')
plt.xlabel('Distance (m)')
plt.ylabel('Speed (km/h)')
plt.axvline(x=1100, color='red', linestyle='--', linewidth=1.5, label='x=2')
plt.axvline(x=1400, color='blue', linestyle=':', linewidth=2)
plt.show()
```

This Graphic can already help us visualize the technicity of this track. Indeed the mix between high and low speed section is a realy triggering factor. By having the track map aside we can already associate specific turn to pick of speed. Such as The low speed pick between 1100 and 1400 meters can be associate with turn 5 to 8.

## Adding Brake and Throttle information to analyse driving style of Max Verstappen. 

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 1, figsize=(20, 8), sharex=True)

axs[0].plot(Fast_lap_VER['Distance'], Fast_lap_VER['Speed'])
axs[0].set_ylabel('Speed (km/h)')

axs[1].plot(Fast_lap_VER['Distance'], Fast_lap_VER['Brake'], color='red')
axs[1].plot(Fast_lap_VER['Distance'], Fast_lap_VER['Throttle']/100, color='green')
axs[1].set_ylabel('Throttle in green (%) and Brake in red (%)')

axs[1].set_xlabel('Distance (m)')

plt.axvline(x=650, color='red', linestyle='--', linewidth=1.5, label='x=2')
plt.axvline(x=850, color='blue', linestyle=':', linewidth=2)

plt.suptitle('Verstappen - Monaco 2023 Qualifying')
plt.show()
```

We can focus ourselves on data between 650 and 850 meters, this section represent turn 3 of the track. As presented by the Brake/Throttle graphic, Max Break 100% when entering the corner to place all the weight on the front wheels, helping the car turning better. Then during the corner he gets progressively on the gas to judge to maximal grip the car.
